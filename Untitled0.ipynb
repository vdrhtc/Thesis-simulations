```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot as plt
from scipy import *
import matplotlib
```

```python
def flux_cubit_potntial(phi, theta, alpha=0.8):
    return 2+alpha-2*cos(phi)*cos(theta) - alpha*cos(1*pi-2*phi)
```

```python
def flux_cubit_potntial1(phi, theta, alpha=0.8):
    return 2+alpha-2*cos(phi)*cos(theta) - alpha*cos(2*pi*0.45+2*phi)
```

```python
font = {'size'   : 17}

matplotlib.rc('font', **font)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
X, Y = linspace(-2*pi, 2*pi, 100), linspace(-2*pi, 2*pi, 100)
XX, YY = meshgrid(X, Y)
Z = []
for n_x in range(len(X)):
    Z.append([])
    for n_y in range(len(Y)):
        Z[n_x].append(flux_cubit_potntial1(X[n_x], Y[n_y], 0.8))
a = axes.contourf(XX/2/pi, YY/2/pi, array(Z), 40, cmap="RdBu_r")
axes.set_aspect(1)
cb = fig.colorbar(a)
cb.set_label("$\mathcal{U}/E_J$", fontsize=25)
# axes.set_title("$Potential\ for\ \\alpha=0.8,\ \Phi = 0.5\Phi_0$", fontsize=20)
scale = 1.5
fig.set_size_inches(fig.get_size_inches()[0]*scale, fig.get_size_inches()[1]*scale)

dict(plt.gca().spines.items())["left"].set_position(('outward', 10))
dict(plt.gca().spines.items())["bottom"].set_position(('outward', 10))
dict(plt.gca().spines.items())["top"].set_position(('outward', 10))
dict(plt.gca().spines.items())["right"].set_position(('outward', 10))
plt.grid()

axes.set_xlabel("$\\theta/2\\pi$", fontsize=25)
axes.set_ylabel("$\\phi/2\\pi$", fontsize=25)

fig.get_size_inches()
```

```python
X = linspace(0, 2*pi, 100)
plt.plot(X/2/pi, [flux_cubit_potntial(x, pi, 1) for x in X], "#000000", ls="-.", label="$\\alpha=1.00$")

plt.plot(X/2/pi, [flux_cubit_potntial(x, pi) for x in X], "black", label="$\\alpha=0.80$")
plt.plot(X/2/pi, [flux_cubit_potntial(x, pi, 0.65) for x in X], "black", ls="--", label="$\\alpha=0.65$")
plt.plot(X/2/pi, [flux_cubit_potntial(x, pi, 0.5) for x in X], "#000000",  ls=":", label="$\\alpha=0.50$")
plt.gcf().set_size_inches(9,7)
# plt.title("$Potential\ at\ \\theta=\pi\ for\ \\Phi=0.5\\Phi_0$", fontsize=20)
plt.legend(loc=0, fontsize=18)
dict(plt.gca().spines.items())["left"].set_position(('outward', 10))
dict(plt.gca().spines.items())["bottom"].set_position(('outward', 10))
dict(plt.gca().spines.items())["top"].set_position(('outward', 10))
dict(plt.gca().spines.items())["right"].set_position(('outward', 10))
plt.grid()

plt.xlabel("$\phi/2\pi$", fontsize=25)
plt.ylabel("$\mathcal{U}/E_J$", fontsize=25)
```

```python
styles = list(reversed([":", "--", "-", "-."]))
for idx, alpha in enumerate(reversed([0.5, 0.65, 0.8, 1])):
    phi_min_1 = arccos(1/2/alpha)
    phi_min_2 = pi-arccos(1/2/alpha)
    theta_min_1 = 0
    theta_min_2 = pi
    k = (phi_min_2 - phi_min_1)/(theta_min_2-theta_min_1)
    b = phi_min_1
    print(phi_min_1/pi, theta_min_1/pi, phi_min_2/pi, theta_min_2/pi)
    U = []
    for theta in linspace(theta_min_1, theta_min_2, 100):
        phi = k*theta+b
        U.append(flux_cubit_potntial(phi, theta, alpha))
    plt.plot(linspace(theta_min_1, theta_min_2, 100)/2/pi, U, "black", ls=styles[idx], label="$\\alpha = %.2f$"%alpha)
plt.gcf().set_size_inches(9,7)
# plt.title("$Potential\ at\ \\theta=\pi\ for\ \\Phi=0.5\\Phi_0$", fontsize=20)
plt.legend(loc=0, fontsize=18)
dict(plt.gca().spines.items())["left"].set_position(('outward', 10))
dict(plt.gca().spines.items())["bottom"].set_position(('outward', 10))
dict(plt.gca().spines.items())["top"].set_position(('outward', 10))
dict(plt.gca().spines.items())["right"].set_position(('outward', 10))
plt.grid()

plt.xlabel("$\\theta/2\pi$", fontsize=25)
# plt.xlabel("$\\theta/2\pi,\ \\phi = (\pi - 2\\cos^{-1} \
#            \\left(\\frac{1}{2\\alpha}\\right))\\theta+\\arccos\,\\frac{1}{2\\alpha}$", fontsize=23)
plt.ylabel("$\mathcal{U}/E_J$", fontsize=25)
```
